### <font color="purple">DATA CLEANING </font>

**_Number of null values in the dataset for every column_**

```python
data.isna().sum()
```

>**<font color="teal">_null_ values</font>**<br><br>
>There are 209 _null_ values in `Year` and `Publisher`.<br><br>
>We can see that no other columns have a single _null_ value
>Also we can easily see from the query below that all the 209 _null_ values of both the columns are same and we need to remove those 209 values only.
<br><br>

```python
data[data['Year'].isna()]
```

>**<font color="teal">_null_ values</font>**<br><br>
>We can see that the rows having _NaN_ in `Year` has also _NaN_ in `Publisher`.<br><br>
>So just eliminate these rows, only if there is a need of data where `Year` or `Publisher` is needed.
>Else, we can use the data without removing _NaN_ or _null_ values.

**_Making a new dataframe for data without null values_**

```python
dataWOna = data[data['Year'].notna()]
print(len(dataWOna), len(data))
dataWOna.reset_index(inplace=True)
print(dataWOna.isna().sum())
dataWOna
```

>**<font color="teal">Removing _null_ values</font>**<br><br>
>On removing all _null_ values we can see that the length of the dataset decreased from **1034** to **825**.<br><br>

### <font color="purple">DATA VISUALISATION</font>

### **_Total number of games published each year_**

```python
years = np.sort(dataWOna['Year'].unique())
gamesEachYr = (dataWOna.groupby('Year').count())['Game'].to_numpy()
print(years, gamesEachYr)
```

>We now have 2 numpy arrays for _years_ and _number of games each year_.
<br><br>

**_Line Plot_**

```python
plt.figure(figsize=(10, 6))

plt.plot(years, gamesEachYr, color='red', linewidth=2, marker='o')
plt.title("Number of Games for PS4 each Year (2013-2020)", y=1.05, fontdict = {'fontsize': 25, 'color': 'blue'})
plt.xlabel("Years", fontdict = {'fontsize': 15}, labelpad=20)
plt.ylabel("Number of Games", fontdict = {'fontsize': 15}, labelpad=30)

plt.grid(linestyle="--", alpha=0.5)
plt.show()
```

**_Bar Plot_**

```python
plt.figure(figsize=(10, 6))

bars = plt.bar(years, gamesEachYr, color=['lightblue', 'cyan', 'orange', 'red', 'red', 'green', 'lightgreen', 'lightgreen'], edgecolor="black")
plt.title("Number of Games for PS4 each Year (2013-2020)", y=1.05, fontdict = {'fontsize': 25, 'color': 'blue'})
plt.xlabel("Years", fontdict = {'fontsize': 15}, labelpad=20)
plt.ylabel("Number of Games", fontdict = {'fontsize': 15}, labelpad=30)

plt.grid(linestyle="--", alpha=0.3)
plt.show()
```

>From both the graphs, we can come to the point that **maximum number of games were released in 2017**.<br><br>
>Also **after 2017, the total number of games substantially decreased**.

### **_Top 10 publishers with Maximum Global Sales PER GAME_**

```python
publishers_Gsales_t10 = publishers_Gsales.copy()

total_games = []
gsales_per_game = []
for i in range(len(publishers_Gsales_t10)):
    pub = publishers_Gsales_t10['Publisher'].iloc[i]
    tot = publisher_data[publisher_data['Publisher']==pub]['Game'].iloc[0]
    total_games.append(tot)
    gsales_per_game.append(publishers_Gsales_t10['Global'].iloc[i]/tot)


publishers_Gsales_t10['Total Games'] = total_games
publishers_Gsales_t10['G-Sales per Game'] = gsales_per_game
publishers_Gsales_t10.sort_values('G-Sales per Game', ascending=False, inplace=True)
publishers_Gsales_t10.head(10)
```

>Created a new dataframe having _publishers_ and their _global sales per game_ with other attributes.
<br>

```python
publishers = publishers_Gsales_t10.head(10)['Publisher'].to_numpy()
gsales_pgame = publishers_Gsales_t10.head(10)['G-Sales per Game'].to_numpy()
```

**_Horizontal Bar Plot_**

```python
plt.figure(figsize=(10, 6))

bars = plt.barh(publishers, gsales_pgame, color=['red', 'orange', 'orange', 'green', 'green', 'green', 'lightgreen', 'lightgreen', 'cyan', 'cyan'], edgecolor="black")
plt.title("Top 10 Publishers with highest \"Global Sales Per Game\" (2013-2020)", y=1.05, fontdict = {'fontsize': 25, 'color': 'blue'})
plt.gca().invert_yaxis()

plt.xlabel("Global Sales Per Game", fontdict = {'fontsize': 15}, labelpad=20)
plt.ylabel("Publishers", fontdict = {'fontsize': 15}, labelpad=15)

plt.grid(linestyle="--", alpha=0.3)
plt.plot()
```

> We can see by the above graph that **Rockstar Games** has significant high Global Sales Per Game till now.<br>
>Followed by **Blizzard Entertainment** and **Warner Bros. Interactive**

### **_Checking Sales in All Regions and Comparing them_**

```python
sales_data = dataWOna.groupby('Year').sum()
sales_data = sales_data[sales_data['Global']!=0.00]
sales_data.reset_index(inplace=True)
sales_data
```

>Created a new dataframe having _years_ and  all the _sales in all regions_.
<br>

```python
na_sales = sales_data['North America']
eu_sales = sales_data['Europe']
jp_sales = sales_data['Japan']
rst_sales = sales_data['Rest of World']
glb_sales = sales_data['Global']

years = sales_data['Year']
```

```python
plt.figure(figsize=(14, 8))

plt.plot(years, na_sales, color='orange', linewidth=3, label='North America')
plt.plot(years, eu_sales, color='blue', linewidth=3, label='Europe')
plt.plot(years, jp_sales, color='green', linewidth=3, label='Japan')
plt.plot(years, rst_sales, color='violet', linewidth=3, label='Rest of World')
plt.plot(years, glb_sales, color='black', linewidth=3, label='Global')


plt.title("Region based Sales comparision each Year (2013-2018)", y=1.05, fontdict = {'fontsize': 25, 'color': 'blue'})
plt.xlabel("Years", fontdict = {'fontsize': 15}, labelpad=20)
plt.ylabel("Sales", fontdict = {'fontsize': 15}, labelpad=30)
plt.legend()
plt.xticks(years)

plt.show()
```

> We can see by the above graph that **Europe** is the biggest market for games over the years as it always has the highest sales as compared to other regions.<br>
>Also, **North America** is the second biggest market for games.
<br>

### **_Correlation between different attributes of the dataset_**

```python
corr_data = data.copy()
corr_data.drop('Year', axis=1, inplace=True)
corr_data = corr_data[corr_data['Global']!=0.00]
corr_mat = corr_data.corr()
corr_mat
```

>**Correlation table to be converted to heatmap**

**_Heat Map_**

```python
plt.figure(figsize=(10, 6))

plt.title("Correlation matrix for Sales across all Regions", y=1.05, fontdict = {'fontsize': 25, 'color': 'blue'})
plt.rc('font', size=12)

sns.heatmap(corr_mat, cmap ="icefire", linewidths = 0.1, annot=True)
```

> We can see by the above graph that **Europe** and **Global** sales have highest correlation.<br>
> And **Japan** and **Europe** sales have least correlation.
<br>
